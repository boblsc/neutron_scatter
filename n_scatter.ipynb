```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,1,100)
```

```python
def E_n(x,m1,m2,E0=2.4):
    gamma=(4*m1*m2)/(m1+m2)**2*x**2
    return E0*(1-gamma)
```

```python
plt.plot(x,np.arcsin(x))
```

```python
def tana(x,m1,m2,E0=2.4):
    b=np.arcsin(x)
    return np.sin(2*b)/(m1/m2-np.cos(2*b))
    
```

```python
plt.plot(x,E_n(x,1,3),label="3m1=m2")
plt.plot(x,E_n(x,1,2),label="2m1=m2")
plt.plot(x,E_n(x,1,1),label="m1=m2")
plt.axhline(0.1,color='grey',linestyle='--')
# plt.axvline(np.cos(0),color='grey',linestyle='--')
plt.xlabel('sin(beta)')
plt.ylabel('cos(alpha)')
plt.legend()
```

```python
for m2 in [0.01,0.1,0.5,1,2,10,100]:
    b=np.linspace(0,np.pi/2,100)
    a=np.arctan(np.sin(2*b)/(1/m2-np.cos(2*b)))
    plt.plot(b,a,label=f'm2={m2}*m1')
    plt.legend()
```

```python
for m2 in [0.01,0.1,0.5,1,2,10,100]:
    b=np.linspace(0,np.pi/2,100)
    a=np.arctan(np.sin(2*b)/(1/m2-np.cos(2*b)))
    plt.plot(b,np.abs(np.sin(a)),label=f'm2={m2}*m1')
    plt.xlabel('cos(beta)')
    plt.ylabel('sin(alpha)')
    plt.legend()
```
